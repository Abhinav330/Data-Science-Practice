## Facebook Friends Netowrk

Data set is called facebook ego dataset Provided by stanford university. link:  https://snap.stanford.edu/data/ego-Facebook.html

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

Reading file

```python
g =nx.read_edgelist('facebook_combined.txt')
```

```python
sp = nx.spring_layout(g)
plt.axis('off')

nx.draw_networkx(g ,sp, with_labels=False,node_size=10)
```

This  is how Network graphs looks like for facebook friends circle

```python
len(g.nodes())
```

### Insight:

```python
len(g.edges())
```

### Insight:
In this Facebook friends network there are total of 88 thousand friends connections (edges)

```python
a =dict(nx.degree(g))
x =max(a, key= lambda x: a[x])
print(x,':',a.get(x))
```

### Insight 
node 107 have the maximum friends that is 1045.

```python
x =min(a, key= lambda x: a[x])
print(x,':',a.get(x))
```

### Insight

```python
edge_list = list(g.edges('11'))
g3 = nx.Graph(edge_list)
nx.draw(g3,with_labels=1,node_size=500, font_size=10)
```

### Insight
Node 11 is only connected to node 0

```python
edge_list = list(g.edges('0'))
print("Total friends of node 0 :", dict(g.degree)['0'])
g2 = nx.Graph(edge_list)
nx.draw(g2,with_labels=1,node_size=500, font_size=10)
```

### Insight
Node 0 have 347 friends. looks like Node 11 may be a fake account of node 0.

```python
i = 0
mi = []
most_influential = nx.degree_centrality(g)

for w in sorted(most_influential, key=most_influential.get, reverse=True):
    print(w, most_influential[w])
    mi.append(w)
    
    if i == 9:
        break
    i=i+1
```

### Top 5 Most connected People in this Facebook friends network

```python
i = 0
most_important_link = nx.eigenvector_centrality(g)

for w in sorted(most_important_link, key=most_important_link.get, reverse=True):
    print(w, most_important_link[w])
    if i == 9:
        break
    i=i+1
```

# TOP 5 Influncers profile on facebook 

Nodes: 1912 , 2266, 2206 , 2233, 2364

#### If not getting much engagement on facebook by some user I will recomment these profiles.

```python
i = 0
connector_links = nx.betweenness_centrality(g)

for w in sorted(connector_links, key=connector_links.get, reverse=True):
    print(w, connector_links[w])
    mi.append(w)
    
    if i == 9:
        break
    i=i+1
```
